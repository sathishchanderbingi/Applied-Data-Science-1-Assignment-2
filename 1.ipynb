```python
temperature= pd.read_csv("temperature.csv")
temperature
```

```python
import matplotlib as plt
global_change= temperature[["Area", "Year", "Value"]].groupby(by="Year").Value.mean()
global_change.plot(kind= "bar",  figsize=(15, 5), title= "Global Temperature Change")
```

```python
global_changes= temperature[["Area", "Year", "Value"]].groupby(by="Year").Value
```

```python
global_change
```

```python
import pandas as pd
from darts import TimeSeries
```

```python
series = TimeSeries.from_series(global_change)
```

```python
series
```

```python
series.plot(label="Change of temperature per year")
#plt.legend(loc= "upper right")
```

```python
train, val = series[:-20], series[-30:]
train
```

```python
#from darts.dataprocessing.transformers import Scaler
```

```python
from darts.models import ExponentialSmoothing

model = ExponentialSmoothing()
model.fit(train)
prediction = model.predict(len(val), num_samples=1000)
```

```python
import matplotlib.pyplot as plt

series.plot()
prediction.plot(label='forecast', low_quantile=0.5, high_quantile=0.5)
plt.legend()
```
